```python
columns=['Sepal_length','Sepal_Width','Petal_lenght','Petal_width','Class_labels']
```

```python
#loading the dataset with column names
df=pd.read_csv("iris.data",names=columns)
df.head()
```

```python
df1=df.values
```

```python
X=df1[:,0:4] #feature varibales
Y=df1[:,4] #target variables
print(Y)
```

```python
import seaborn as sb
```

```python
sb.pairplot(df,hue="Class_labels")
```

```python
#splitting the model
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
```

```python
#logistic regression
from sklearn.linear_model import LogisticRegression
```

```python
model1=LogisticRegression()
```

```python
model1.fit(X_train,Y_train)
```

```python
model1.score(X_test,Y_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_predicted=model1.predict(X_test)
print(accuracy_score(Y_test,y_predicted))
```

```python
from sklearn.metrics import classification_report
print(classification_report(Y_test,y_predicted))
```

So by using logistic regression and training the model the trained model is giving 97 percent of accuracy

```python
#now we will train by using  decision tree classifier
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model2=DecisionTreeClassifier()
model2.fit(X_train,Y_train)
```

```python
model2.score(X_test,Y_test)
```

```python
y_predicted2=model2.predict(X_test)
print(accuracy_score(Y_test,y_predicted2))
print(classification_report(Y_test,y_predicted2))
```

```python
#now we will train using support vector machine
```

```python
from sklearn.svm import SVC
```

```python
model3=SVC()
model3.fit(X_train,Y_train)
model3.score(X_test,Y_test)
```

```python
y_predicted3=model3.predict(X_test)
print("The accuracy score is",accuracy_score(Y_test,y_predicted3))
      
```

```python
print("The classification report is",classification_report(Y_test,y_predicted3))
```
